Условие задачи:

f(x) = sin(x)^2 - cos(x)^2

1. Определить корни
2. Найти интервалы, на которых функция возрастает
3. Найти интервалы, на которых функция убывает
4. Построить график
5. Вычислить вершину
6. Определить промежутки, на котором f > 0
7. Определить промежутки, на котором f < 0

Решение:

4. Построение графика

```python
from sympy.plotting import plot
from sympy import Symbol, solve, diff, evalf, solveset, S, pprint, sin, cos

x = Symbol('x')
f = sin(x)**2 - cos(x)**2
p1 = plot(f, show=False)
p1.show()
```

1. Определение корней.

```python
"""Решение в аналитическом виде"""
equat_solut = solve(f,x)
print(equat_solut)
"""Решение в численном виде"""     
for i in equat_solut:
    print(i.evalf())
```

5. Вычисление вершины

```python
diff_f = diff(f,x)
print(diff_f)
solve(diff_f,x)
```

2. Нахождение интервалов, на которых функция возрастает

```python
pprint(solveset(diff_f > 0, x, domain=S.Reals), use_unicode=True)
```

3. Нахождение интервалов, на которых функция убывает

```python
pprint(solveset(diff_f < 0, x, domain=S.Reals), use_unicode=True)
```

6. Определение промежутков, на которых f > 0

```python
pprint(solveset(f > 0, x, domain=S.Reals), use_unicode=True)
```

7. Определение промежутков, на которых f < 0

```python
pprint(solveset(f < 0, x, domain=S.Reals), use_unicode=True)
```
